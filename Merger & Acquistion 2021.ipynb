<h4> Acquisitions through the Years</h4>

```python
#Acquisition Date' is in datetime format
df['Acquisition Year'] = pd.to_datetime(df['Acquisition Year'], errors='coerce')
```

```python
#Extract the year from 'Acquisition Date'
df['Acquisition Year'] = df['Acquisition Year'].dt.year
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("ticks")

plt.figure(figsize=(10, 10))
sns.histplot(
    data=df,
    x="Acquisition Year",
    bins=25,
    discrete=True,
    kde=False
)
plt.title("Acquisitions Over the Years")
plt.xlabel("Year")
plt.ylabel("Count")
sns.despine()
plt.show()
```

<h4>What are the most popular businesses to acquire?</h4>

```python
# Calculate the frequency of each acquired business
acquired_business_counts = df['Business'].value_counts().head(20) # Get the top 10

# Create a bar chart of the top 10 acquired businesses
plt.figure(figsize=(10, 6))
acquired_business_counts.plot(kind='bar')
plt.title('Most Popular Acquired Businesses (Top 20)')
plt.xlabel('Business')
plt.ylabel('Number of Acquisitions')
plt.xticks(rotation=45, ha='right')
plt.show()

# Print the table of the top 10 acquired businesses
print(acquired_business_counts.to_markdown(numalign="left", stralign="left"))
```

<h4>Most common types of businesses acquired</h4>

```python
# Calculate the frequency of each acquired business
acquired_business_counts = df['Business'].value_counts().head(20)

# Create a bar chart of the most common acquired businesses
plt.figure(figsize=(10, 6))
acquired_business_counts.plot(kind='bar')
plt.title('Most Common Types of Businesses Acquired (Top 20)')
plt.xlabel('Business Type')
plt.ylabel('Number of Acquisitions')
plt.xticks(rotation=45, ha='right')
plt.show()

# Print the table of the most common acquired businesses
print(acquired_business_counts.to_markdown(numalign="left", stralign="left"))
```

<h4>Where are the acquisitions of companies concentrated?</h4>

```python
# Calculate the frequency of acquisitions by country
acquisitions_by_country = df['Country'].value_counts().head(10).reset_index()
acquisitions_by_country.columns = ['Country', 'Number of Acquisitions']

# Create a bar chart of the top 10 countries using seaborn
plt.figure(figsize=(10, 6))

sns.barplot(x='Country', y='Number of Acquisitions', data=acquisitions_by_country)
plt.title('Acquisitions by Country (Top 10)')
plt.xlabel('Country')
plt.ylabel('Number of Acquisitions')
plt.xticks(rotation=45, ha='right')
plt.show()

# Print the table of the top 10 countries
print(acquisitions_by_country.to_markdown(index=False, numalign="left", stralign="left"))
```

<h4>What does each company spend on average on an acquisition?</h4>

```python
# Calculate the average acquisition price for each parent company
average_acquisition_price = df.groupby('Parent Company')['Acquisition Price'].mean().sort_values(ascending=False).head(10) #keep top 10

# Create a bar chart
plt.figure(figsize=(10, 6))
average_acquisition_price.plot(kind='bar')
plt.title('Average Acquisition Price per Parent Company (Top 10)')
plt.xlabel('Parent Company')
plt.ylabel('Average Acquisition Price')
plt.xticks(rotation=45, ha='right')
plt.show()

# Print the result
print("Average Acquisition Price per Company:")
print(average_acquisition_price.to_markdown(numalign="left", stralign="left"))
```

<h4>Which products were derived from a high number of company acquisitions?

```python
# Calculate the frequency of each derived product
product_counts = df['Derived Products'].value_counts().sort_values(ascending=False).head(10).reset_index()
product_counts.columns = ['Derived Products', 'Count']

# Create the bar chart using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Derived Products', y='Count', data=product_counts)
plt.title('Products Derived from a High Number of Company Acquisitions (Top 10)')
plt.xlabel('Derived Products')
plt.ylabel('Number of Acquisitions')
plt.xticks(rotation=45, ha='right')
plt.show()

# Print the results
print("Products Derived from a High Number of Company Acquisitions:")
print(product_counts.to_markdown(index=False, numalign="left", stralign="left"))
```

<h4> Which years had the highest number of acquisitions?</h4>

```python
# Make 'Acquisition Year' to be numeric and not NaN
df['Acquisition Year'] = pd.to_numeric(df['Acquisition Year'], errors='coerce')
df_filtered = df.dropna(subset=['Acquisition Year'])  # removes any rows where Acquisition Year is NaN using df.dropna()

# Calculate the number of acquisitions per year
acquisitions_per_year = df_filtered['Acquisition Year'].value_counts().sort_values(ascending=False).head(15).reset_index()
acquisitions_per_year.columns = ['Acquisition Year', 'Number of Acquisitions'] # The columns are renamed to Acquisition Year and Number of Acquisitions for use with Seaborn.

# Create the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Acquisition Year', y='Number of Acquisitions', data=acquisitions_per_year)
plt.title('Top 15 Years with Highest Number of Acquisitions')
plt.xlabel('Acquisition Year')
plt.ylabel('Number of Acquisitions')
plt.xticks(rotation=45, ha='right')
plt.show()

# Print the years with the highest number of acquisitions
print("Years with the highest number of acquisitions:")
print(acquisitions_per_year.to_markdown(index=False, numalign="left", stralign="left"))
```
